```python
data_set= {
    'no of hours' : [1, 2, 5, 7, 8, 10, 12, 15, 17],
    'age' : [20, 34, 21, 27, 34, 21, 20, 45, 31],
    'salary' : [1000, 3000, 5000, 8000, 8500, 9000, 12000, 15000, 22000]
}
```

```python
import pandas as pd
data_filed = pd.DataFrame(data_set)
print(data_filed)
```

```python
import matplotlib.pyplot as plt
import seaborn as sb
sb.scatterplot(x="no of hours",y="age",data= data_filed)
plt.xlabel("no of hours ")
plt.ylabel("age")
plt.title("no of houre vs age")
plt.show()
```

```python

import matplotlib.pyplot as plt
import seaborn as sb
sb.scatterplot(x="salary",y="age",data= data_filed)
plt.xlabel("salary")
plt.ylabel("age")
plt.title(" age vs salary ")
plt.show()
```

```python
sb.scatterplot(x="no of hours",y="salary",data= data_filed)
plt.xlabel("no of hours")
plt.ylabel("salary")
plt.title(" no of hours vs salary ")
plt.show()
```

```python
X=data_filed.iloc[:,:2]
y=data_filed.iloc[:,:-1]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
lr = LinearRegression()
model = lr.fit(X_train, y_train)
pred = model.predict(X_test)
print(pred)
```

```python
print("Intercept :",model.intercept_)
```

```python
print("Slope :", model.coef_)
```
